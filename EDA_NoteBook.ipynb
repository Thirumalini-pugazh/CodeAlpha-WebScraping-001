```python
import pandas as pd

data = {
    "Order_ID":[1001,1002,1003,1004,1005,1006,1007,1008,1009,1010],
    "Category":["Electronics","Furniture","Clothing","Electronics","Furniture","Clothing","Electronics","Furniture","Clothing","Electronics"],
    "Sales":[25000,15000,8000,30000,12000,9000,28000,18000,11000,35000]
}

df = pd.DataFrame(data)
df.head()
```

```python
df.info()
df.describe()
df["Sales"].sum()
df["Sales"].mean()
df["Category"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Sales"])
plt.show()
```

```python
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.show()
```
